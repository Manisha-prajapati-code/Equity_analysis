# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
file_path = 'Equity_Analyst_Dataset.xlsx'  # Make sure this file is in the same directory
data = pd.read_excel(file_path)

# Display the first few rows of the dataset
print("Dataset Preview:")
display(data.head())
```

# Set a consistent style for visualizations


```python
sns.set_style("whitegrid")
```

# 1. Market Cap Distribution by Sector

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sector', y='Market Cap (in Cr)', data=data, palette='viridis')
plt.xticks(rotation=45)
plt.title('Market Cap Distribution by Sector', fontsize=14)
plt.xlabel('Sector', fontsize=12)
plt.ylabel('Market Cap (in Cr)', fontsize=12)
plt.show()
```

# 2. P/E Ratio vs. Current Price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='P/E Ratio', y='Current Price (₹)', hue='Sector', data=data, palette='coolwarm')
plt.title('P/E Ratio vs. Current Price', fontsize=14)
plt.xlabel('P/E Ratio', fontsize=12)
plt.ylabel('Current Price (₹)', fontsize=12)
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# 3. Dividend Yield by Sector

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sector', y='Dividend Yield (%)', data=data, ci=None, palette='muted')
plt.xticks(rotation=45)
plt.title('Average Dividend Yield by Sector', fontsize=14)
plt.xlabel('Sector', fontsize=12)
plt.ylabel('Dividend Yield (%)', fontsize=12)
plt.show()
```

# 4. Volume Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Volume (in Lakhs)'], kde=True, bins=30, color='purple')
plt.title('Volume Distribution', fontsize=14)
plt.xlabel('Volume (in Lakhs)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

# 5. 1-Year Return by Sector

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sector', y='1-Year Return (%)', data=data, palette='pastel')
plt.xticks(rotation=45)
plt.title('1-Year Return Distribution by Sector', fontsize=14)
plt.xlabel('Sector', fontsize=12)
plt.ylabel('1-Year Return (%)', fontsize=12)
plt.show()
```

# 6. Top 10 Companies by Market Cap

```python
top_10 = data.nlargest(10, 'Market Cap (in Cr)')
plt.figure(figsize=(10, 6))
sns.barplot(x='Market Cap (in Cr)', y='Company Name', data=top_10, palette='flare')
plt.title('Top 10 Companies by Market Cap', fontsize=14)
plt.xlabel('Market Cap (in Cr)', fontsize=12)
plt.ylabel('Company Name', fontsize=12)
plt.show()
```
